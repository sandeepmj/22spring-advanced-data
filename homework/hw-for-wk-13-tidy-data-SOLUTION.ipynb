### Melt Parameters

```id_vars``` – This is the column(s) that we want to unpivot around so values should show up more than once.

```value_vars``` – All columns that are not ```id_vars``` will unpivot. Named ```variable``` and ```value``` by default.

```var_name``` – We can name the column variable header(s).

```value_name``` – We can name the result column(s).



````python
## read data file
df = pd.read_csv("messy/census.csv")
df
````

````python
## I don't feel like typing out column names so I create a list
list(df.columns)
````

````python
## The columns I want to unpivot around are the first 5
## Create an id_vars list by slicing the columns list
id_vars_list = list(df.columns)[0:5]
id_vars_list
````

````python
## melt your df using the id_vars_list
## NOTE: in this cell I did not specify the varible and value names

dfm = df.melt(id_vars = id_vars_list)
dfm
````

````python
## let's specify the variable and value column headers
dfm = df.melt(id_vars = id_vars_list,
             var_name = "year",
             value_name = "amount")
dfm
````

````python
## Note that amount is a string
dfm.info()
````

````python
## remove the thousand separators 
dfm["amount"] = dfm["amount"].str.replace(",","")
````

````python
dfm
````

````python
## The thousand separators are gone, but it's still a string object
dfm.info()
````

````python
dfm["amount"] = pd.to_numeric(dfm["amount"], errors="coerce")
dfm.info()
````
