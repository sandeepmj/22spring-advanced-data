## Read two or more separate files and combine them.

Often, we have to work with ```bike_data_2019.csv```, ```bike_data_2020.csv```, etc.

But if we really want to see change over time, we need to combine these spreadsheets into a single spreadsheet.

Bring in ```flt_delays_17.csv``` and ```flt_delays_18.csv``` and combine them into a dataframe called ```delay_df```

````python
## first read the individual csv files and store in dataframes
flt17_df = pd.read_csv('class6-data/flt_delays_17.csv')
flt18_df = pd.read_csv('class6-data/flt_delays_18.csv')
````

````python
## confirm they are in similar columns

flt17_df.info()
````

````python
## If we only want to see how many rows and columns
flt17_df.shape
````

````python
## confirm they are in similar columns

flt18_df.info()
````

````python
## If we only want to see how many rows and columns

flt18_df.shape
````

````python
## create a list of column headers
headers18 = list(flt18_df.columns)
headers18
````

````python
## create a list of the other
headers17 = list(flt17_df.columns)
headers17
````

````python
## confirm they are identical
headers17 == headers18
````

````python
## concat the two by placing both dataframes in a list
delay_df = pd.concat([flt17_df,flt18_df])
````

````python
## How many total rows?

delay_df.shape
````

````python
## what type of data is a shape
type(delay_df.shape)
````

````python
## we can pull out values from a tuple, just like a list or dictionary
delay_df.shape[0]
````

````python
## the number of rows should add up correctly
delay_df.shape[0] == flt17_df.shape[0] + flt18_df.shape[0]
````

## Programmatically creating a single df from multiple files

In the previous section, we individually imported flight delay data for each year. What if we had 20 years (or more) of flight delay data. We'll want to do it programmatically.



````python
## Glob is a package that "globs" target files into a list
import glob
````

````python
## glob all files that start with "flt" and end in ".csv". 

my_files = glob.glob("class6-data/flt*.csv")
my_files
````

````python
## Use for loop
multi_dfs = [] ## initialize an emtpy list

for my_file in my_files:
    multi_dfs.append(pd.read_csv(my_file)) ## append each csv that is read into a df into our list
    
````

````python
## note that if we call an item in our list, we see each of our dataframes
multi_dfs[0]
````

````python
## concat the list of dataframes into a single data
final_df = pd.concat(multi_dfs) 
````

````python
## a sample shows us dates from each year.
final_df.sample(20)
````

````python
## confirm number of columns etc.
final_df.info()
````
