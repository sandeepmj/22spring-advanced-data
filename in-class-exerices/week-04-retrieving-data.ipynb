# Retrieving data

## retrieving data from lists...

### 1. List basics:

````python
## Let's build a list with the following terms:

toxic_chemicals = [
    "Polyvinyl Chloride",
    "PVC",
    "Perfluorooctanoic Acid",
    "PFOA",
    "Triclosan",
    "Bisphenol-A",
    "BPA",
    "Formaldehyde",
    "Naphthalene",
    "Asbestos"
]
````

````python
## call the list
toxic_chemicals
````

````python
## print the list
print(toxic_chemicals)
````

````python
## how long is this list?
len(toxic_chemicals)
````

````python
## what type of object?
type(toxic_chemicals)
````

### 2. Slicing to target items in a list

````python
## Last time we used index() method to retrieve 
## the index position of an item:

toxic_chemicals.index("PVC")
````

#### Now we want to retrieve the data, not the index

#### Now we want to retrieve the data, not the index

```list_name[index_position]```

````python
## retrieve the data in the zeroth position

toxic_chemicals[0]
````

````python
## call the second item in the list

toxic_chemicals[1]
````

````python
## call the last item in the list
toxic_chemicals[-1]
````

````python
## call the penultimate item in the list
toxic_chemicals[-2]
````

#### How about retrieving a subset of data and not just individual data points?

#### Slice notation

Screen Shot 2022-02-24 at 4.57.54 PM.png

````python
## Call the first 3 items:
toxic_chemicals[0:2]
````

#### Slice notation

Screen Shot 2022-02-24 at 5.01.32 PM.png

````python
## Call the first 3 items:
toxic_chemicals[0:3]
````

````python
## RUN THIS CELL - 
### Use this list of CEO salaries from 1985 
ceo_salaries_1985 = [150_000, 201_000, 110_000, 75_000, 92_000, 55_000]
ceo_salaries_1985
````

````python
## Now update each salary to 2021 dollars.
## Print the following info:
## "A CEO's salary of [1985 salary] in 1985 is worth [updated salary] in 2021 dollars."
cpi_1985 = 	107.6
cpi_2021 = 271.4
for salary in ceo_salaries_1985:
    updated_sal = (salary/cpi_1985) * cpi_2021
    print(f"A CEO's salary of ${salary} in 1985 is worth ${updated_sal:,.0f} in 2021 dollars.")
````

````python
sal_2021 = []
cpi_1985 = 	107.6
cpi_2021 = 271.4
for salary in ceo_salaries_1985:
    updated_sal = (salary/cpi_1985) * cpi_2021
    sal_2021.append(int(round(updated_sal,0)))
    
sal_2021
````
